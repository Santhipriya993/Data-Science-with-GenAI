### Data Cleaning

```python
df['DOJ']=pd.to_datetime(df['DOJ'])
df['DOL'].replace(to_replace='present',value=pd.Timestamp.now(),inplace=True)
df['DOL']=pd.to_datetime(df['DOL'])
```

```python
df['workperiod'] = (df['DOL']-df['DOJ'])/np.timedelta64(1, 'W')
df['workperiod'] = df['workperiod'].astype(int)
df['workperiod'].unique()
```

Note : Here there are some negative values , the reason might be offer rejections.

```python
from datetime import datetime, date  

# CREATING AGE COLUMN USING 'DOB' COLUMN
df['DOB'] = pd.to_datetime(df['DOB'])

current_date = pd.to_datetime(datetime.now())
df['Age'] = (current_date - df['DOB']).dt.days / 365.25  # Divide by 365.25 for average years
df['Age'] = df['Age'].astype(int)  # Convert to integer for whole years
df['Age']
```

```python
#ROWS WITH GRADUATION YEAR AS ZERO
pd.set_option('display.max_columns',None)
df[df['GraduationYear']==0]
```

There is only one row with graduation year as ‘0’. It can be replaced with 2014 since the 12th graduation year is 2010 and DOJ is 2014.

```python
df['GraduationYear'].replace(to_replace = 0, value=2014,inplace = True)
```

#### Note : JobCity column has -1 value so need to replace with 'others'

```python
df['JobCity'].replace(to_replace = -1, value='others', inplace=True)
```

#### Is there a relationship between gender and specialization? (i.e. Does the preference of Specialisation depend on the Gender?)

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

# Your existing code for creating the cross-tab and stacked bar plot
cross_tab = pd.crosstab(df['Gender'], df['Specialization'])
ax = cross_tab.plot(kind='bar', stacked=True, figsize=(10, 10))

# Adjusting the legend to be outside the plot
ax.legend(title='Specialization', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.title('Gender vs Specialization')
plt.xlabel('Gender')
plt.ylabel('Count')

# Adjust layout to make room for the legend
plt.tight_layout()

# Show the plot
plt.show()
```

The plot shows that the majority of individuals in both genders are enrolled in electronics and communication engineering. However, there are some differences in the distribution of specializations between genders. Males are more likely to be enrolled in mechanical engineering, computer science and engineering, and information technology, while females are more likely to be enrolled in electronics and electrical engineering, instrumentation and control engineering, and applied electronics and instrumentation engineering.

This suggests that there may be gender-based preferences or biases in the choice of specialization, and further analysis would be needed to understand the underlying reasons for these differences.

```python
# Optional: Perform a chi-square test for independence
chi2, p, dof, expected = chi2_contingency(cross_tab)
print(f"Chi-square statistic: {chi2}, p-value: {p}")
```

**Chi-square Statistic**: A chi-square statistic of approximately 104.47 indicates a strong deviation from the expected frequencies. This suggests that the observed distribution of gender across different specializations is significantly different from what would be expected if there were no association between gender and specialization.

**P-value:** The p-value of approximately 1.25e-06 (or 
0.000001245
0.000001245) is extremely small, which is well below the common significance level of 0.05. This indicates strong evidence against the null hypothesis (which states that there is no association between gender and specialization).

# RESEARCH QUESTION

### Higher Education leads to better wages.

```python
# Converting Bachelors to UnderGrad and Masters to PostGrad
df['Grad_level'] = ['UnderGrad' if x == 'B.Tech/B.E.' else 'PostGrad' for x in df['Degree']]

# Using T-test to check the hypothesis
from scipy import stats

undergrad_sal = df.loc[df['Grad_level']=='UnderGrad','Salary']
postgrad_sal = df.loc[df['Grad_level']=='PostGrad','Salary']


t_statistic, p = stats.ttest_ind(undergrad_sal, postgrad_sal)
print("p-value:", p)

if p < 0.05:
    print("Reject the null hypothesis. There is a statistically significant difference in the average salary between undergraduates and postgraduates.")
else:
    print("Fail to reject the null hypothesis. There is no statistically significant difference in the average salary between undergraduates and postgraduates.")
```
